```python
files = [f for f in os.listdir('cleaned_data')]
```

```python
import time
import re
```

```python
files[0]
```

```python
dfs = []
for file in files:
    start = time.time()
    df = pd.read_parquet(f'cleaned_data/{file}')
    new_df = (df.review_date.str[:4] + '-' + df.review_date.str[5:7]).reset_index().groupby('review_date').count()
    new_df = new_df.reset_index().rename(columns={'index':'count'})
    new_df.review_date = pd.to_datetime(new_df.review_date)
    name = re.findall('(?<=amazon_reviews_)[.\w]*(?=.parq)', file)[0].lower()
    new_df['product_category'] = name
    dfs.append(new_df)
    print(name, new_df.shape, f'{(time.time()-start)/60:,.2f} minutes')
```

```python
df = pd.concat(dfs)
```

```python
fig = px.line(
    df.sort_values(by='review_date'), 
    x='review_date', 
    y='count', 
    labels={
        'count':'Review Count',
        'review_date':'Monthly Reviews',
        'product_category':'Product Category'},
    color='product_category',
    color_discrete_sequence=cc.glasbey[:37],
    title='Monthly Reviews Over Time By Product Category',
    width=1200,
    height=920)
fig.update_layout(
    font_family='Open Sans',
    font_size=20,
    legend_font_size=12,
    legend_title_font_size=20
)
fig.show()
```

```python
fig = px.line(
    df.sort_values(by='review_date'), 
    x='review_date', 
    y='count', 
    labels={
        'count':'Review Count (Log Scaled)',
        'review_date':'Monthly Reviews',
        'product_category':'Product Category'},
    log_y=True,
    color='product_category',
    color_discrete_sequence=cc.glasbey[:37],
    title='Monthly Reviews Over Time By Product Category',
    width=1200,
    height=920)
fig.update_layout(
    font_family='Open Sans',
    font_size=20,
    legend_font_size=12,
    legend_title_font_size=20
)
fig.show()
```
